### Histograms
Histograms convey below details about the image
1. Contrast
2. How bright or How dark
3. Different pixel intensities & their dominance


gray scale = pixel values =  0 - 255 ( values close to 0 means black(darker) and 255 mean white(brighter))


# Create a histogram
new function: cv2.calcHist([input image],[channels],masking,number of bins,range of scale)

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread("messi2.png")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("gray image",gray)


#create histogram
hist = cv2.calcHist([gray],[0],None,[256],[0,255])
plt.figure()
plt.xlim([0,255])
plt.xlabel("Pixel intensities")
plt.ylabel("Number of pixel's")
plt.plot(hist, color='r')
plt.show()         


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("messi2.png")

b_hist = cv2.calcHist([img],[0],None,[256],[0,256])
g_hist = cv2.calcHist([img],[1],None,[256],[0,256])
r_hist = cv2.calcHist([img],[2],None,[256],[0,256])

plt.figure()
plt.xlabel("Intensity levels")
plt.ylabel("number of pixels")
plt.plot(b_hist,color='b'),plt.xlim([-20,300])
plt.plot(g_hist,color='g'),plt.xlim([-20,300])
plt.plot(r_hist,color='r'),plt.xlim([-20,300])
plt.show()  
```

### Snipping Utility

```python
import cv2

coordinates = [] # coordinates of image we want to crop
status = False # status of whether we are cropping or not

# lets create a function - A MOUSE CALL BACK FUNCTION
#whener we press left mouse button or release the button, this function will be called
# helps in cropping the image we want to crop

def lets_crop(event,x,y,flags,parameters):# should be in same order
    
    global coordinates,status # making them global variables
    # when user clicks left mouse button, status is true and we start recording the intial point coordiantes
    if event == cv2.EVENT_LBUTTONDOWN:
        status = True
        coordinates = [(x,y)]# record the intial coordinates
    ## when user clicks left mouse button, status is true and we stop cropping and record the final point coordiantes    
    if event == cv2.EVENT_LBUTTONUP:
        status = False
        coordinates.append((x,y))# append the final coordinates to coordinates variable list
        #To draw the rectangle for the cropped image
        cv2.rectangle(image,coordinates[0],coordinates[1],(0,255,0),2)
        # display the cropped image
        cv2.imshow("Crop",image)
    
image = cv2.imread("cactus.png")
# we are perform the cropping function on the copy image because if we perform
# it on original, the change will be permananet on original there by we wont 
# be having an extra image to compare
copy = image.copy()
# create an empty window to display cropped image
cv2.namedWindow("Crop")
# seeting the call back function
#Crop is the window we are going to use to display cropped image 
cv2.setMouseCallback("Crop",lets_crop)

# if user wants to redo cropping
while True:
    #display the image again
    cv2.imshow("Crop",image)
    keypress = cv2.waitKey(1)
    # user needs to press r to redo cropping
    if keypress == ord('r'):
        image = copy.copy()
    # user needs to press c to confirm with existing one
    if keypress == ord('c'):
        break

# to display cropped image in to a new window
if len(coordinates) == 2: # intial and final points
    roi = copy[coordinates[0][1]:coordinates[1][1], coordinates[0][0]:coordinates[1][0]]
    cv2.imshow("Region Of Interest", roi)
    cv2.waitKey(0)

cv2.destroyAllWindows()
    
```

### Coding exercise, Writing our own thersholding program

```python
import cv2
import numpy as np
import random

image = cv2.imread("tj.jpg")
```

```python
cv2.imshow("original",image)

# convert to gray scale image
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#take the value of rows and cols in to row and cols valriables like below
(rows,cols) = np.shape(gray)
#set a threshold value
thresh = 140

# rather than using cv2.threshold function, let's use manual code to set a threshold
for i in range(0,rows):
    for j in range(0,cols):
        if gray[i][j] >=thresh:
            gray[i][j] = 255
        else:
            gray[i][j] = 0

cv2.imshow("Threshold image",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Adaptive thresholding
types of methods:<br>
cv2.ADAPTIVE_THRESH_GAUSSIAN_C<br>
cv2.ADAPTIVE_THRESH_MEAN_C

```python
import cv2
import numpy as np
import random

image = cv2.imread("tj.jpg")
```

```python
cv2.imshow("original",image)

# convert to gray scale image
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#Since the image is not as expected, there are few false positive
# so let's smooth the image before feeding in to threshold function
gray = cv2.GaussianBlur(gray,(3,3),0)

ret, thresh_global = cv2.threshold(gray,140,255,cv2.THRESH_BINARY)
# adaptive threshold function
# instead of supplying manula threshold value, we assign a adaptive threshold function
thresh_adaptive_Mean = cv2.adaptiveThreshold(
    gray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,31,3
    )
thresh_adaptive_gaussian = cv2.adaptiveThreshold(
    gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,33,1
    )

#cv2.imshow("thresh_global",thresh_global)
cv2.imshow("thresh_adaptive_Mean",thresh_adaptive_Mean)
cv2.imshow("thresh_adaptive_gaussian",thresh_adaptive_gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Custom Glittering Filter
Extention to adaptive threshold filter

```python
import cv2
import numpy as np
import random

image = cv2.imread("people.jpg")
```

```python
cv2.imshow("original",image)

# convert to gray scale image
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# so let's smooth the image before feeding in to threshold function
gray = cv2.GaussianBlur(gray,(7,7),0)

thresh_adaptive_gaussian = cv2.adaptiveThreshold(
    gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,1
    )

(rows,cols)= np.shape(thresh_adaptive_gaussian)
canvas = np.zeros((rows,cols,3),dtype = "uint8")
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            #if the pixel valus != 0 (black), then
            if thresh_adaptive_gaussian[i][j] != 0:
                #take a random color value
                color = random.randrange(0,255,1)
                # and replace it with that pixel value
                canvas[i][j][k] = color
            else:
                #is the pixel value is black == 0
                color =0
                #then leave it as it is ( balck )
                canvas[i][j][k] = color
                

cv2.imshow("GLittered Image",canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Changing color spaces in the video

```python
import cv2
import numpy as np

capture = cv2.VideoCapture(0)
# 0  - connects to internal webcan
# 1 - connects to external cams 
flag = 0 # 0 - RGB, 1 - Gray, 2 - HSV

# in order to capture eachh frame every second and display continously in the
# form of video, we need to loop the process
while (True):
    
    ret, frame = capture.read()# each frame recorded from the 
    #video will be stored ijn frame variable
    
    keypress = cv2.waitKey(1)
    if keypress == ord('r'):
        flag = 0
    # turns RGB scale
    if keypress == ord('g'):
        flag = 1
    # turns GRAY scale
    if keypress == ord('h'):
        flag = 2
    # turns HSV scale
    if flag == 0:
        pass
    if flag == 1:
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if flag == 2:
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    cv2.imshow("Video Color spaces", frame)
    
    if keypress == 27:
        break
capture.release()
cv2.destroyAllWindows()
```

## Object tracking using its color

The problem with tracking an object using its color is "ILLUMINATION"
because, based on illumination, the pixel value may be effected in the video capture there by effecting the tracking process.

```python
import cv2
import numpy as np

video = cv2.VideoCapture(0)

while True:
    ret, frame = video.read()
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    # range of color values
    low_range = np.array([110,50,50])
    up_range = np.array([130,255,255])
     
    mask = cv2.inRange(hsv, low_range, up_range)
    cv2.imshow("masked",mask)
    
    keypress = cv2.waitKey(1)
    if keypress == 27:
        break
video.release()
cv2.destroyAllWindows()
```

## Countours
detecting and drawing countours

```python
import cv2
import numpy as np
image = cv2.imread("contour.jpg")

def filter_my_contours(image,contours):
    filter_contours = []
    ImageArea = image.shape[0] * image.shape[1]
    for i in contours:
        # criteria for filtering
        area = cv2.contourArea(i)
        if (area >= 0.10 * ImageArea) and (area <= 0.70 * ImageArea):
            
            filter_contours.append(i)
    return filter_contours
```

```python
blurred = cv2.pyrMeanShiftFiltering(image,120,180)# removes all the noise in the image
gray = cv2.cvtColor(blurred,cv2.COLOR_BGR2GRAY)
# OTSU's Binarization link below
#https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_thresholding/py_thresholding.html
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
_, contours, _ = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
print(len(contours))

#to remove false contours
filter_contours = filter_my_contours(image,contours)
print(len(filter_contours))

# Draw the contours
#cv2.drawContours(inout image,contours,number of contours to display
#                , color of contour,thickness)
cv2.drawContours(image,filter_contours,-1,(0,0,255),2)# -1 = draw out all contours



cv2.namedWindow("Display", cv2.WINDOW_NORMAL)
cv2.imshow("Display",image)
cv2.imshow("Thresh",thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### how to draw bounding box for the image

```python
import cv2
import numpy as np
image = cv2.imread("contour4.jpg")

def filter_my_contours(image,contours):
    filter_contours = []
    ImageArea = image.shape[0] * image.shape[1]
    for i in contours:
        # criteria for filtering
        area = cv2.contourArea(i)
        if (area >= 0.1 * ImageArea) and (area <= 0.90 * ImageArea):
            
            filter_contours.append(i)
    return filter_contours
```

```python
blurred = cv2.pyrMeanShiftFiltering(image,151,200)# removes all the noise in the image
gray = cv2.cvtColor(blurred,cv2.COLOR_BGR2GRAY)
# OTSU's Binarization link below
#https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_thresholding/py_thresholding.html
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
_, contours, _ = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
print(len(contours))

#to remove false contours
filter_contours = filter_my_contours(image,contours)
#draw the bounding box
#take the shappe value
(x,y,w,h) = cv2.boundingRect(filter_contours[0])
image = cv2.rectangle(image,(x,y),(x+w,y+h), (255,0,0),2)
cropped = image[y:y+h,x:x+w]

print(len(filter_contours))

# Draw the contours
#cv2.drawContours(inout image,contours,number of contours to display
#                , color of contour,thickness)
cv2.drawContours(image,filter_contours,-1,(0,0,255),2)# -1 = draw out all contours



cv2.namedWindow("Display", cv2.WINDOW_NORMAL)
cv2.imshow("Display",image)
cv2.imshow("Thresh",thresh)
cv2.imshow("cropped",cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Masking - tocapture exactly image shape with out any backgroud 
minimun bouding function will be used

```python
import cv2
import numpy as np
image = cv2.imread("contour7.jpg")

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
blurred = cv2.medianBlur(gray,27)
ret, thresh = cv2.threshold(blurred,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
#cv2.bitwise_image(input image, input image, mask method(thresh value))
masked_image = cv2.bitwise_and(image,image, mask = thresh)

cv2.namedWindow("Display", cv2.WINDOW_NORMAL)
cv2.imshow("Display",image)
cv2.imshow("thresh",thresh)
cv2.imshow("masked",masked_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
